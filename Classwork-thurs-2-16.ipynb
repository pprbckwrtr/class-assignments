```python
def makeBirdDict(filename):
    # this function opens the data files from Thomas et al and extracts the species and families into a dictionary
    birdDict ={}
    fobj = open(filename,"r", encoding = "ISO-8859-15") #open the files
    all_records = fobj.readlines () [1:] #read the file
    for line in all_records:
        record_elements = line.split(",")
        sp = record_elements[0]
        fam = record_elements[2]
        birdDict[sp]=fam
        #split lines on comma 
    #stick species and family into birdDict
    fobj.close()
    return birdDict
makeBirdDict("/home/eeb177-student/Desktop/eeb-177/homework/formatted.nature.csv")
```
